Q1.Ask the user to enter a value (e.g., M) for a mathematical function.
 Generate x values from -10 to 10 using np.linspace().
 Compute y values for:
o y= M⋅x2
o y=M⋅sin(x)
 Plot both functions in a single figure:
o Use different colors and line styles.
o Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as plt
M = float(input("Enter a value for M: "))
x = np.linspace(-10, 10, 100)
y1 = M * x**2
y2 = M * np.sin(x)

# Plotting both functions in a single figure
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label=r'$y = M \cdot x^2$', color='b', linestyle='--')
plt.plot(x, y2, label=r'$y = M \cdot sin(x)$', color='r', linestyle='-')
plt.title(f"Plot of Functions for M = {M}")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

Q2.Create a dataset of five subjects and scores.
o Convert it into a Pandas DataFrame.
o Plot the scores using a Seaborn bar plot with:
    o Different colors for each bar.
    o Annotations on top of each bar.
    o Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    'Subject': ['Math', 'Science', 'English', 'Computer', 'Economics'],
    'Score': [85, 90, 74, 68, 92]}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
barplot = sns.barplot(x='Subject', y='Score', data=df, palette='Set2')

for p in barplot.patches:
    barplot.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='center', fontsize=12, color='black', xytext=(0, 8), textcoords='offset points')
plt.title('Scores by Subject', fontsize=16)
plt.xlabel('Subject', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.grid(True)
plt.show()
```

Q3.Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().
 Create a 2x2 subplot layout displaying:
o Line plot showing cumulative sum.
o Scatter plot with random noise.
 Customize with titles, axis labels, and grids.

```python
import numpy as np
import matplotlib.pyplot as plt
roll_number = 102317213
np.random.seed(roll_number)

data = np.random.randn(50)

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].plot(np.cumsum(data), label='Cumulative Sum', color='b')
axs[0].set_title('Cumulative Sum of Values')
axs[0].set_xlabel('Index')
axs[0].set_ylabel('Cumulative Sum')
axs[0].grid(True)
axs[0].legend()

noise = np.random.randn(50) * 0.5
axs[1].scatter(range(50), data + noise, label='Data with Noise', color='r')
axs[1].set_title('Scatter Plot with Random Noise')
axs[1].set_xlabel('Index')
axs[1].set_ylabel('Values with Noise')
axs[1].grid(True)
axs[1].legend()
plt.tight_layout()
plt.show()
```

Q4.Download Data-set from the below link
https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv
Apply ‘seaborn’ library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
df = pd.read_csv(url)

plt.figure(figsize=(8, 5))
sns.lineplot(data=df, x='month_number', y='total_profit', marker='o')
plt.title("Total Profit Over Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
for column in df.columns[1:-1]:
    plt.plot(df['month_number'], df[column], marker='o', label=column)
plt.title("Product Sales Over Time")
plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
totals = df.sum(numeric_only=True)
totals.drop(['month_number'])
sns.barplot(x=totals.index, y=totals.values, palette='Set3')
plt.title("Total Sales and Profit by Category")
plt.xlabel("Features")
plt.ylabel("Total")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
