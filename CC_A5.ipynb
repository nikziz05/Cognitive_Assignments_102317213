Q1.For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), Find
i. Sum of all elements
ii. Sum of all elements row-wise
iii. Sum of all elements column-wise

```python
import numpy as np
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

sum_all_elements = np.sum(gfg)
print(f"Sum of all elements: {sum_all_elements}")

sum_row_wise = np.sum(gfg, axis=1)
print(f"Sum of elements row-wise: {sum_row_wise}")

sum_column_wise = np.sum(gfg, axis=0)
print(f"Sum of elements column-wise: {sum_column_wise}")
```

Q2. (a)For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find
i. Sorted array
ii. Indices of sorted array
iii. 4 smallest elements
iv. 5 largest elements
(b) For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find
i. Integer elements only
ii. Float elements only

```python
#a)
import numpy as np
array = np.array([10, 52, 62, 16, 16, 54, 453])

sorted_array = np.sort(array)
print(f"Sorted array: {sorted_array}")

sorted_indices = np.argsort(array)
print(f"Indices of sorted array: {sorted_indices}")

smallest_4_elements = np.partition(array, 4)[:4]
print(f"4 smallest elements: {smallest_4_elements}")

largest_5_elements = np.partition(array, -5)[-5:]
print(f"5 largest elements: {largest_5_elements}")
#b)
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

integer_elements = array[array == array.astype(int)]
print(f"Integer elements only: {integer_elements}")

float_elements = array[array != array.astype(int)]
print(f"Float elements only: {float_elements}")
```

Q3.You are given a weekly sales dataset and need to perform various operations
using NumPy broadcasting.
a) Generate your unique sales dataset:
 Take the sum of the ASCII values of the initials of your first and last
name. Call this value X. (If your initials are A B → ASCII sum = 65
+ 66 = 131 → sales = [131, 181, 231, 281, 331].)
 Create a NumPy array sales with values [X, X+50, X+100, X+150,
X+200].
b) Compute your personalized tax rate as ((X % 5) + 5) / 100.
 Use broadcasting to apply this tax rate to each sales value.
c) Adjust sales based on discount:
 If sales < X+100, apply a 5% discount.
 If sales >= X+100, apply a 10% discount.
d) Expand sales data for multiple weeks:
 Create a 3×5 matrix representing three weeks of sales by stacking
sales three times using broadcasting.
 Increase sales by 2% per week using element-wise broadcasting.

```python
import numpy as np
X = ord('A') + ord('B')

sales = np.array([X, X+50, X+100, X+150, X+200])
print(f"Sales array: {sales}")

tax_rate = ((X % 5) + 5) / 100
print(f"Tax rate: {tax_rate}")

sales_with_tax = sales * (1 + tax_rate)
print(f"Sales after applying tax: {sales_with_tax}")

discounted_sales = np.where(sales < X + 100, sales * 0.95, sales * 0.90)
print(f"Sales after applying discount: {discounted_sales}")

sales_weekly = np.tile(sales, (3, 1))
print(f"Sales data for three weeks (before increase):\n{sales_weekly}")

sales_weekly = sales_weekly * (1 + 0.02) ** np.arange(3)[:, None]
print(f"Sales data for three weeks (after 2% weekly increase):\n{sales_weekly}")
```

Q4.Generate x values using np.linspace() from -10 to 10 with 100 points. Use
each function from the list below and compute y values using NumPy:
 Y = x2
 Y = sin(x)
 Y = ex
 Y = log(|x| + 1)
Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)
y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)

plt.figure(figsize=(10, 8))

#Plot Y = x^2
plt.subplot(2, 2, 1)
plt.plot(x, y1, label=r'$Y = x^2$', color='b')
plt.title(r'Plot of $Y = x^2$')
plt.xlabel('x')
plt.ylabel('Y')
plt.grid(True)
plt.legend()

#Plot Y = sin(x)
plt.subplot(2, 2, 2)
plt.plot(x, y2, label=r'$Y = sin(x)$', color='g')
plt.title(r'Plot of $Y = sin(x)$')
plt.xlabel('x')
plt.ylabel('Y')
plt.grid(True)
plt.legend()

#Plot Y = e^x
plt.subplot(2, 2, 3)
plt.plot(x, y3, label=r'$Y = e^x$', color='r')
plt.title(r'Plot of $Y = e^x$')
plt.xlabel('x')
plt.ylabel('Y')
plt.grid(True)
plt.legend()

#Plot Y = log(|x| + 1)
plt.subplot(2, 2, 4)
plt.plot(x, y4, label=r'$Y = log(|x| + 1)$', color='orange')
plt.title(r'Plot of $Y = log(|x| + 1)$')
plt.xlabel('x')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
