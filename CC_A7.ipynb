```python
# PART I: Randomized Sales Data Generation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Initialize Random Seed
roll_number = 102317213
np.random.seed(roll_number)

# 2. Generate Sales Data
sales_data = np.random.randint(1000, 5000, size=(12, 4))

# 3. Convert to DataFrame
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
df = pd.DataFrame(sales_data, columns=categories, index=months)
df

# PART II: Data Manipulation and Analysis
# 1. First 5 rows and summary stats
print(df.head())
print(df.describe())

# 2. Total sales per category and per month
total_by_category = df.sum()
total_by_month = df.sum(axis=1)

# 3. Average sales growth between consecutive months per category
avg_growth = df.pct_change().mean()

# 4. Add Total Sales and Growth Rate columns
df['Total Sales'] = df.sum(axis=1)
df['Growth Rate'] = df['Total Sales'].pct_change() * 100
df

# 5. Apply Discount
if roll_number % 2 == 0:
    df['Electronics'] *= 0.9
else:
    df['Clothing'] *= 0.85
df
# PART III: Visualization
# 1. Monthly sales trends (line plot)
df[categories].plot(figsize=(10, 5), marker='o')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.grid()
plt.legend()
plt.show()

# 2. Box plots for each category
plt.figure(figsize=(8, 5))
sns.boxplot(data=df[categories])
plt.title('Sales Distribution by Category')
plt.grid()
plt.show()
```

Q2.For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
i. Find element-wise absolute value
ii. Find the 25th, 50th, and 75th percenƟle of flattened array, for each column, for each
row.
iii. Mean, Median and Standard DeviaƟon of flattened array, of each column, and
each row

```python
import numpy as np
array = np.array([[1, -2, 3], [-4, 5, -6]])
abs_array = np.abs(array)
print("Element-wise Absolute Value:\n", abs_array)

flattened = array.flatten()
print("\nPercentiles (Flattened Array):")
print("25th:", np.percentile(flattened, 25))
print("50th (Median):", np.percentile(flattened, 50))
print("75th:", np.percentile(flattened, 75))

print("\nPercentiles (Column-wise):")
print("25th:", np.percentile(array, 25, axis=0))
print("50th:", np.percentile(array, 50, axis=0))
print("75th:", np.percentile(array, 75, axis=0))

print("\nPercentiles (Row-wise):")
print("25th:", np.percentile(array, 25, axis=1))
print("50th:", np.percentile(array, 50, axis=1))
print("75th:", np.percentile(array, 75, axis=1))

print("\nStatistics (Flattened Array):")
print("Mean:", np.mean(flattened))
print("Median:", np.median(flattened))
print("Std Dev:", np.std(flattened))

print("\nStatistics (Column-wise):")
print("Mean:", np.mean(array, axis=0))
print("Median:", np.median(array, axis=0))
print("Std Dev:", np.std(array, axis=0))

print("\nStatistics (Row-wise):")
print("Mean:", np.mean(array, axis=1))
print("Median:", np.median(array, axis=1))
print("Std Dev:", np.std(array, axis=1))
```

Q3.For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]).
i. Find floor, ceiling and truncated value, rounded values

```python
import numpy as np
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
floors = np.floor(a)
ceils = np.ceil(a)
truncs = np.trunc(a)
rounded = np.round(a)

print("Original Array:", a)
print("Floor Values:", floors)
print("Ceiling Values:", ceils)
print("Truncated Values:", truncs)
print("Rounded Values:", rounded)
```

Q4.WAP to swap two elements in a list in Python Using a Temporary Variable


```python
def swap_elements(list1, index1, index2):
    print("Original List:", list1)
    temp = list1[index1]
    list1[index1] = list1[index2]
    list1[index2] = temp
    print(f"List after swapping elements at index {index1} and {index2}:", list1)

my_list = [10, 20, 30, 40, 50]
swap_elements(my_list,1,3)
```

Q5.Write a Python program to swap two elements in a set by converting it to a list first.

```python
def swap_set_elements(s, val1, val2):
    print("Original Set:", s)
    list1 = list(s)
    try:
        idx1 = list1.index(val1)
        idx2 = list1.index(val2)
        temp = list1[idx1]
        list1[idx1] = list1[idx2]
        list1[idx2] = temp
        new_set = set(list1)
        print(f"Set after swapping {val1} and {val2}:", new_set)
    except ValueError as e:
        print("One or both values not found in the set:", e)
my_set = {1,2,3,4,5}
swap_set_elements(my_set,2,5)
```
